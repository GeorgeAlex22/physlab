```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy as sp
import scipy.optimize as opt
from physlab import curvefit

```

```python
# genetrate fake data
n = 20
g = 9.806
v0 = 12.22
y0 = 0.33
st=0.1
sy = 0.5
t = np.linspace(0,2,n)
y = y0 + v0*t - 0.5*g*t**2 + np.random.randn(n)*sy
sy = sy + np.random.randn(n)*sy/10
st = st + np.random.randn(n)*st/10
# define model function
def model(t, a, b, c):
    return a*t*t+b*t+c
# apply curve fit
params, cov = curvefit(model,data_x = t, data_y = y, data_sx = st, data_sy = sy, xlabel = 'time (s)', ylabel = 'height (m)', latexModel = '\\alpha_0 t^2 +\\alpha_1 t + \\alpha_2 ', title = 'curvefit Example')
```
